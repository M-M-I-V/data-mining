```python
import pandas as pd

data = {
    'CustomerID': ['001', '002', '003', '004'],
    'Age': [25, None, 35, 29],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Anual Income': [40000, 48000, None, 52000],
    'Total Purchases': [5, 8, 6, 9],
    'CLV': [1500, 2000, 1800, 2200]
}

df = pd.DataFrame(data)
print('Original data:\n', df)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = 'mean')
df[['Age', 'Anual Income']] = imputer.fit_transform(df[['Age', 'Anual Income']])
df
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df
```

```python
scaler = MinMaxScaler()
df[['Age', 'Anual Income', 'Total Purchases']] = scaler.fit_transform(df[['Age', 'Anual Income', 'Total Purchases']])
df
```

```python
df_final = df.drop('CustomerID', axis = 1)
print('Preprocessed Data: \n', df_final)
```

```python
x = df_final.drop('CLV', axis = 1)
y = df_final['CLV']

print('Features: \n', x)
print('\nTarget: \n', y)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data = {
    'CustomerID': ['001', '002', '003', '004'],
    'Age': [25, None, 35, 29],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Anual Income': [40000, 48000, None, 52000],
    'Total Purchases': [5, 8, 6, 9],
    'CLV': [1500, 2000, 1800, 2200]
}

df = pd.DataFrame(data)
print('Original data:\n', df)

imputer = SimpleImputer(strategy = 'mean')
df[['Age', 'Anual Income']] = imputer.fit_transform(df[['Age', 'Anual Income']])
df

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df

scaler = MinMaxScaler()
df[['Age', 'Anual Income', 'Total Purchases']] = scaler.fit_transform(df[['Age', 'Anual Income', 'Total Purchases']])
df

df_final = df.drop('CustomerID', axis = 1)
print('\nPreprocessed Data: \n', df_final)

x = df_final.drop('CLV', axis = 1)
y = df_final['CLV']

print('\nFeatures: \n', x)
print('\nTarget: \n', y)
```
