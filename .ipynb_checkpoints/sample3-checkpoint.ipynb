```python
import pandas as pd

data = {
    'StudentID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Gender': ['Male', 'Female', 'Female', 'Male', None],
    'Math Score': [85, 90, None, 75, 88],
    'English Score': [78, 85, 80, None, 82],
    'Passed': ['Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
print('Original Data: \n', df)
```

```python
from sklearn.impute import SimpleImputer

score_imputer = SimpleImputer(strategy = 'mean')
df[['Math Score', 'English Score']] = score_imputer.fit_transform(df[['Math Score', 'English Score']])
df
```

```python
gender_imputer = SimpleImputer(strategy = 'most_frequent')
df[['Gender']] = gender_imputer.fit_transform(df[['Gender']])
df
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Passed'] = encoder.fit_transform(df['Passed'])
df
```

```python
scaler = StandardScaler()
df[['Math Score', 'English Score']] = scaler.fit_transform(df[['Math Score', 'English Score']])
df
```

```python
df_cleaned = df.drop('StudentID', axis = 1)
print('\nCleaned and Processed Data: \n', df_cleaned)
```
