```python
import pandas as pd

data = {
    'CustomerID': ['C001', 'C002', 'C003', 'C004'],
    'Age': [25, None, 40, 35],
    'Gender': ['Male', 'Female', 'Female', None],
    'Satisfaction': ['High', 'Low', 'Medium', 'High'],
    'Spending Score': [60, 40, None, 75]
}

df = pd.DataFrame(data)
print('Original Data: \n', df)
```

```python
from sklearn.impute import SimpleImputer

num_imputer = SimpleImputer(strategy = 'mean')
df[['Age', 'Spending Score']] = num_imputer.fit_transform(df[['Age', 'Spending Score']])
df
```

```python
cat_imputer = SimpleImputer(strategy = 'most_frequent')
df[['Gender']] = cat_imputer.fit_transform(df[['Gender']])
df
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender']) # 1 - Male, 0 - Female
df['Satisfaction'] = encoder.fit_transform(df['Satisfaction']) # 0 - High, 1 - Low, 2 - Medium
df
```

```python
scaler = MinMaxScaler()
df[['Age', 'Spending Score']] = scaler.fit_transform(df[['Age', 'Spending Score']])
df
```

```python
df_cleaned = df.drop('CustomerID', axis = 1)
print('\nCleaned and Processed Data: \n', df_cleaned)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data = {
    'CustomerID': ['C001', 'C002', 'C003', 'C004'],
    'Age': [25, None, 40, 35],
    'Gender': ['Male', 'Female', 'Female', None],
    'Satisfaction': ['High', 'Low', 'Medium', 'High'],
    'Spending Score': [60, 40, None, 75]
}

df = pd.DataFrame(data)
print('Original Data: \n', df)

num_imputer = SimpleImputer(strategy = 'mean')
df[['Age', 'Spending Score']] = num_imputer.fit_transform(df[['Age', 'Spending Score']])
df

cat_imputer = SimpleImputer(strategy = 'most_frequent')
df[['Gender']] = cat_imputer.fit_transform(df[['Gender']])
df

encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Satisfaction'] = encoder.fit_transform(df['Satisfaction'])
df

scaler = MinMaxScaler()
df[['Age', 'Spending Score']] = scaler.fit_transform(df[['Age', 'Spending Score']])
df

df_cleaned = df.drop('CustomerID', axis = 1)
print('\nCleaned and Processed Data: \n', df_cleaned)
```
